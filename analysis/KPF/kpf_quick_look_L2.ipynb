```python
data_dir = "/Users/hisaacson/code/Intro-2-Spectra/data/kpf/"
f1 = 'KP.20230103.71033.03_L2.fits'  # This is an LFC file, so the CC shows many peaks.
f1 = 'KP.20230114.24430.58_L2.fits' # updated file from jan 14th commissioning night.
print(data_dir+f1)
hdu1=fits.open(data_dir+f1)
hdu_hd = hdu1[0].header
hdu1.info()
print(repr(hdu_hd))
```

```python
# The primary data products are the cross correlation curves
l2g_ccf = hdu1['GREEN_CCF'].data
l2r_ccf = hdu1['RED_CCF'].data
print('Dimensions of L1 file: ',l2g_ccf.shape,l2r_ccf.shape)
```
