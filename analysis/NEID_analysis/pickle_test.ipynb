```python
key1 = 'order'
key2 = 'x'
key3 = 'y' 

c = 0
test_dict = {key1:[],key2:[],key3:[]}
for i in np.arange(0,10):
    test_dict[key1].append(i)
    test_dict[key2].append([i,i+1,i+2])
    test_dict[key3].append([i,i-1,i-2])
    
   
```

```python
test_dict
```

```python
import pandas as pd
```

```python
test_data = pd.DataFrame(test_dict)
```

```python
order_data = test_data['order']
x_data = test_data['x']
y_data = test_data['y']
```

```python
import matplotlib.pyplot as plt

for i in np.arange(len(x_data)):
    plt.plot(x_data[i],y_data[i], label = str(order_data[i]))
    plt.title('testing this feature lol')
    plt.legend()

plt.show()
             
                   
```
